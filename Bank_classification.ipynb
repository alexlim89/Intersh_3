### Categorical variables

```python
categorical = X_train[['job','marital','education','default','housing','loan','contact','month','poutcome']]
categorical.describe()
```

```python
# по категориальным переменным: "job"(профессия):12 уникальных значений чаще всего это менеджемент, частота 175.541 раз
# "marital"(семейной положение) : 3 уникальных значений, чаще всего в браке, частота 480759
# "education"(образование): 4 уникальных значений чаще всего среднее , 401683 раз
# "default"(есть ли не выплаченные кредиты): 2 значения , чаше всего нет 737151 раз
# "housing"(есть ли ипотека): 2 значения, чаще всего да, 411288 раз
# "loan"(есть ли потребительский кредит): 2 значения, чаще всего нет , 645023 раз
# "contact"(тип контакта): 3 значения , чаще всего cellular, 486655 раз
# "mounth"(месяц): 12 значений и чаще всего месяц май 228411 раз
# "poutcome"(результат предыдущей компании):4 значения чаше всего "unknown" 672450 раз
```

```python
def plot_category_by_target(data, category_col, target_col, figsize=(12,5), hue_order=(0,1)):
    
    # Порядок категорий: возьмём по убыванию count при y=1 (можешь сменить на 0 при желании)
    order = (data.loc[data[target_col] == hue_order[-1], category_col]
                 .value_counts()
                 .index)
    if len(order) == 0:
        order = data[category_col].value_counts().index

    # Считаем кросстаб по количествам и тоталы по классам
    counts = pd.crosstab(data[category_col], data[target_col])
    counts = counts.reindex(index=order, columns=hue_order, fill_value=0)
    class_totals = counts.sum(axis=0)  # сколько всего объектов в каждом классе y

    plt.figure(figsize=figsize)
    ax = sns.countplot(
        data=data,
        x=category_col,
        hue=target_col,
        order=order,
        hue_order=hue_order
    )

    # Аннотируем: для каждого столбика (bar) считаем долю внутри его класса
    for h_idx, container in enumerate(ax.containers):
        y_class = hue_order[h_idx]
        total_in_class = class_totals[y_class] if class_totals[y_class] > 0 else 1
        for bar_idx, bar in enumerate(container):
            height = bar.get_height()  # Count(category, y)
            if height <= 0:
                continue
            cat = order[bar_idx]
            pct = counts.loc[cat, y_class] / total_in_class * 100
            ax.annotate(f'{pct:.1f}%',
                        (bar.get_x() + bar.get_width()/2, height),
                        ha='center', va='bottom', fontsize=9)

    plt.title(f'{category_col} share within each {target_col} (P({category_col} | y))')
    plt.xlabel(category_col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

### Job

```python
plot_category_by_target(X_train,'job','y', hue_order=(0,1))
```
